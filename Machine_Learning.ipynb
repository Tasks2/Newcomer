STEP12: EVALUATE THE ALGORITHM

```python
from sklearn import metrics
import numpy as np
mae=metrics.mean_absolute_error(y_test, y_pred)
mse=metrics.mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
print('MEAN ABSOLUTE ERROR (MAE):', mae)
print('MEAN SQUARED ERROR(MSE)', mse)
print('ROOT MEAN SQUARED ERROR (RMSE)', rmse)
```

HOW TO SAVE THE LINEAR REGRESSION MODEL

```python
import joblib
import pickle
from sklearn.linear_model import LinearRegression
#Train your LinearRegression Model
model=LinearRegression()
model.fit(x,y)#Replace X and y with your training data
#save model using joblib
joblib.dump(model,'linear_regression_model.pkl')
#or save model using pickle
with open('linear_regression_model.pkl', 'wb') as file:
  pickle.dump(model,file)
```

USE TRAINED MODEL TO TEST MORE DATASET

```python
import joblib
import numpy as np
import pandas as pd
#Sample x values for prediction
new_X=np.array([6,7,8,9,10]).reshape(-1,1)#reshape to a 2d array
model=joblib.load('linear_regression_model.pkl')
#make predictions on the new X values
predictions=model.predict(new_X)
#Create a DataFrame with one-dimensional arrays
new_data=pd.DataFrame({'X': new_X.flatten(), 'Predicted_Y': predictions.flatten()})
#display the new_data Data Frame with the X values and the predicted Y values
print(new_data)
```
